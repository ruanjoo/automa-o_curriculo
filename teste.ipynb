```python
import pyautogui
import time
import pyperclip
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
```

```python
driver = webdriver.Firefox()
wait = WebDriverWait(driver, 10)
driver.get("https://sidia.inhire.app/sidia/vagas")
driver.maximize_window()

barra_pesquisa = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.ID, "name"))
)
barra_pesquisa.clear()
barra_pesquisa.send_keys("está" + Keys.ENTER)
barra_pesquisa.send_keys(Keys.TAB)  # Move o foco para o próximo elemento

time.sleep(5)  # Aguarda a página carregar
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(3)  # Aguarda carregar o conteúdo após o scroll


pyautogui.hotkey('ctrl', 'a')  # Seleciona tudo
time.sleep(1)
pyautogui.hotkey('ctrl', 'c')  # Copia o conteúdo selecionado
time.sleep(1)

conteudo = pyperclip.paste()  # Pega o conteúdo da área de transferência

from IPython.display import display  # Adicione esta importação no início do notebook

linhas = conteudo.split('\n')

dados = [linha.split('\t') for linha in linhas if linha.strip() != '']

df = pd.DataFrame(dados)
  # Mostra a tabela formatada no Jupyter

def filtrar_estagios(linhas):
    palavras_chave = ['estágio', 'estagiario']
    return [linha for linha in linhas if any(palavra in linha.lower() for palavra in palavras_chave)]

linhas = conteudo.split('\n')
linhas_estagio = filtrar_estagios(linhas)

# Cria DataFrame apenas com as linhas filtradas (cada linha vira uma linha da tabela)
df_estagios = pd.DataFrame(linhas_estagio, columns=["Vagas de Estágio"])

# Exibe somente as vagas filtradas em formato de tabela
display(df_estagios)
vagas_tecnologia = df_estagios["Vagas de Estágio"].tolist()
#fechar o navegador
driver.quit()
```
